# 데이터 불러오기 

```python
train = pd.read_csv('C:/Users/bsh36/Desktop/Data/benz/train.csv')
test = pd.read_csv('C:/Users/bsh36/Desktop/Data/benz/test.csv')
sub = pd.read_csv('C:/Users/bsh36/Desktop/Data/benz/sample_submission.csv')
```

```python
pd.options.display.max_columns=999
```

```python
alldata = pd.concat([train,test])
alldata.head()
```

```python
alldata2 = alldata.drop(columns=['y'])
alldata2.head()
```

```python
# 문자형 to 숫자형
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
c = alldata2.columns[alldata2.dtypes==object]
for i in c:
    alldata2[i] = le.fit_transform(alldata2[i])
alldata2
```

```python
train2 = alldata2[:len(train)]
test2 = alldata2[len(train):]
```

```python
# 데이터가 복잡하면 depth를 6이상, 아닌경우 depth 6이하
# 데이터의 컬럼을 다 사용하는것 colsample_bytree = 1과 일부만 사용하기 0~1 
# 컬럼의 개수가 많으면 colsample_bytree = 0.8은 높다..
# n_estimators 나무생성, 딥러닝의 epochs
# learning_rate 속도가 느려지면 학습량을 많이 trade-off 관계 
from xgboost import XGBRegressor
xgb = XGBRegressor(learning_rate = 0.01, max_depth =3, colsample_bytree = 0.5,n_estimators = 800)
```

```python
from catboost import CatBoostRegressor
# 캣부스트는 학습률을 간접, 직접 조정할수있음
# iteration =1500하니 성능이 낮아짐.
cbc = CatBoostRegressor(verbose=50,depth=3, learning_rate = 0.008,iterations=1000)
```

```python
xgb.fit(train2, train['y'])
# 캣부스트는 범주형 컬럼을 설정해주면 좋음
cbc.fit(train2, train['y'],cat_features = [1,2,3,4,5,6,7,8])
```

```python
# 앙상블 
result = xgb.predict(test2)
result1 = cbc.predict(test2)
```

```python
# 앙상블중 가장 기본적인 블렌딩. 가중치만 정의해주기
sub['y'] = result *0.8 + result1*0.2 
sub.to_csv('C:/Users/bsh36/Desktop/Data/benz/ensemble.csv',index=False)
```
