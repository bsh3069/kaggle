```python
train = pd.read_csv('C:/Users/bsh36/Desktop/Data/house/train.csv')
test = pd.read_csv('C:/Users/bsh36/Desktop/Data/house/test.csv')
sub = pd.read_csv('C:/Users/bsh36/Desktop/Data/house/sample_submission.csv')
```

# 전처리

```python
alldata = pd.concat([train,test])
# 집의 연식 컬럼 생성
# 도움이 되는 컬럼으로만 
alldata['old'] = alldata['YrSold']-alldata['YearBuilt']
alldata['total'] = alldata['GrLivArea'] + alldata['TotalBsmtSF']
alldata['quality']= alldata['OverallQual'] +  alldata['OverallCond']
alldata['final_area'] = alldata['total'] + alldata['GarageArea']
# 원래 있는 컬럼 심화 시키기
alldata['overall_minus'] = alldata['OverallQual']-alldata['OverallCond']
alldata['area']= alldata['final_area'] + alldata['LotArea']*0.1
alldata
```

```python
# 상관관계 확인
alldata.corr()['SalePrice'].sort_values(ascending=False)
```

```python
alldata2 = alldata.drop(columns=['Id','SalePrice'])
```

```python
# 문자형 컬럼들을 원핫 인코딩
alldata2['MSSubClass'] = alldata2['MSSubClass'].astype(object)
alldata3 = pd.get_dummies(alldata2)
alldata3
```

```python
# 결측치 처리
alldata3 = alldata3.fillna(-1)
```

```python
from sklearn.preprocessing import StandardScaler, RobustScaler
scaler = StandardScaler()
# 중앙값을 가지고 스케일링 - 이상치에 덜 민감함
rb = RobustScaler()
```

```python
alldata4 = rb.fit_transform(alldata3)
alldata4 = pd.DataFrame(alldata4, columns = alldata3.columns)
alldata4
```

```python
train2 = alldata4[:len(train)]
test2 = alldata4[len(train):]
```

# 모델링

```python
from sklearn.linear_model import Ridge
r2 = Ridge(alpha=14)
```

```python
from catboost import CatBoostRegressor
cbc = CatBoostRegressor(verbose=1000,iterations=4000)
```

```python
r2.fit(train2, np.log(train['SalePrice']))
cbc.fit(train2,np.log(train['SalePrice']))
```

```python
result = r2.predict(test2)
result2 = cbc.predict(test2)
```

```python
# 블렌딩
sub['SalePrice'] = np.exp(result2) *0.5 + np.exp(result)* 0.5 
```

```python
sub.to_csv('C:/Users/bsh36/Desktop/Data/house/cbc.csv',index=False)
```
